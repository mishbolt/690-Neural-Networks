## Taylor's Theorem
Let $f: \mathbb{R}\to \mathbb{R}$ be a $k+1$ times differentiable at the point $a\in\mathbb{R}$ for some nonnegative integer $k$. Then for all $x\in \mathbb{R}$ there exists $\xi$ between $a$ and $x$, such that

$$f(x)=P_k(x)+R_k(x)$$ 

where

$$P_k(x)=\sum_{i=0}^k \frac{f^{(i)}(a)}{i!}(x-a)^i,$$

and

$$R_k(x)=\frac{f^{(k+1)}(\xi)}{(i+1)!}(x-a)^{i+1}.$$

## Plot
We plot Gaussian curves for different values of $\sigma$,centered at $\mu=0$:
$$f(x)=\frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{1}{2}(\frac{x-\mu}{\sigma})^{\frac{1}{2}}}$$



```python
import matplotlib.pyplot as plt
import numpy as np

a= np.arange(-5,5.5,0.05)

for sig in [2,1.5,1,0.5]:
    plt.plot(a,1/(sig*(2*np.pi)**(1/2))*np.exp(-1/2*(a/sig)**2))
    plt.title('Gaussian (Bell) curve, $\sigma=[2,1.5,1,0.5], \mu=0$')
```
