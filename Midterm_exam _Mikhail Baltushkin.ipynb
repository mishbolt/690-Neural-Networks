Here is the chain of implications:
\begin{align*}
\text{Differentiability+Bounded Derivative}\overset{4}{\implies}\text{Liepschitz}\overset{1}{\implies}\text{Uniform Continuity}\overset{2}{\implies}\text{Continuity}\\
\sideset{\scriptsize{3}}{}\Uparrow\\
\text{Differentiability}
\end{align*}
In all of the proofs, $f:\mathbb{R}\to\mathbb{R}$.

### Proof of 1:

Suppose $f$ is K-Liepschitz. Let $\varepsilon>0$, and let $\delta = \dfrac{\varepsilon}{K}>0$. Then, if $|x-y| < \delta$ for any $x,y\in \mathbb{R}$, we have:
$$|f(x)-f(y)|<K|x-y|<K\delta=\varepsilon.$$
Thus $f$ is uniformly continuous.

---

### Proof of 2:

Trivially true

### Proof of 3:
Suppose $f$ is differentiable at $x_0$. Then for all $\varepsilon>0$ there exists $\delta>0$ and a number $D\in\mathbb{R}$ such that if $|x_0-y|<\delta$, then $\left|\dfrac{f(x_0)-f(y)}{x_0-y}-D\right|<\varepsilon$, i.e. $$\lim_{y\to x_0}\dfrac{f(x_0)-f(y)}{x_0-y}=D$$
Now observe that for $x_0\neq y$, we have $$f(x_0)-f(y) = \dfrac{f(x_0)-f(y)}{x_0-y}( x_0-y),$$ hence the limit:
$$\lim_{y\to x_0}f(x_0)-f(y) = \lim_{y\to x_0}\dfrac{f(x_0)-f(y)}{x_0-y}\cdot\lim_{y\to x_0}(x_0-y) = D\cdot 0=0.$$
Thus, we have $$\lim_{y\to x_0}f(x_0)-f(y)=0,$$ i.e. for any $\varepsilon>0$ there exists $\delta>0$ such that if $|x_0-y|<\delta$, $|f(x_0)-f(y)|<\varepsilon$, therefore $f$ is continuous at $x_0$

---

### Proof of 4:
Suppose $f$ is differentiable and $|f'(x)|\leq K$ for all $x\in \mathbb{R}$ and some $K>0$. Suppose $x,y\in \mathbb{R}$, then by the Mean-Value Theorem there exists $c\in \mathbb{R}$ such that

$$f(x)-f(y)=f'(c)(x-y)\implies |f(x)-f(y)|=|f'(c)||x-y|\leq K|x-y|$$

Hence $f$ is K-Liepchitz

### Uniform continuity doesn't imply Liepshitz

Let $f:\mathbb{R}^+\to \mathbb{R}$ defined via $f(x)=\sqrt{x}$. For any $n\in \mathbb{N}$, $|f(\frac{1}{n})-f(0)|=|\frac{1}{\sqrt{n}}|=\sqrt{n}|\frac{1}{n}-0|$, so there is no M satisfying the requirement for liepschitz 

---

### Continuity doesn't imply uniform continuity

Let $f:(0,1] \to \mathbb{R}$ be defined via $f(x)=\frac{1}{x}$, which is continuous on $(0,1]$. \
Also suppose $f$ is uniformly continuous. Then for any $\varepsilon>0$ such that if $x,y\in (0,1]$ with $|x-y|<\delta$, then $|f(x)-f(y)|<\varepsilon$. Let $\varepsilon=\frac{1}{2}$. Then we can choose $\delta>\dfrac{1}{N}$ for some $N>1$. Then for $x = \dfrac{1}{N},y=\dfrac{1}{N+1}\in (0,1]$, $$|x-y|=\left|\frac{1}{N}-\frac{1}{N+1}\right|=\left|\frac{1}{N(N+1)}\right|<\left|\frac{1}{N}\right|<\delta$$ however:
$$|f(x)-f(y)|=N+1-N=1>\varepsilon$$
which is a contradiction.


---

### Continuity doesn't imply differentiability

Consider $f:\mathbb{R}\to \mathbb{R}$ defined via 
$f(x)=
\begin{cases}
x & x>0\\
0 & x\leq0
\end{cases}$

Then $f$ is continuous on $\mathbb{R}$, however the limit from the left
$$\lim_{y\to 0^{-}}\frac{f(y)-f(0)}{y-0}=0$$

and the limit from the right:
$$\lim_{y\to 0^{+}}\frac{f(y)-f(0)}{y-0}=\lim_{y\to 0^{+}}\frac{y}{y}=1$$

disagree hence $f$ is not differentiable at 0.

---

### Differentiability doesn't imply uniform continuity

We use the same example $f:(0,1] \to \mathbb{R}$ defined via $f(x)=\frac{1}{x}$ is differentiable with the derivative being $f'(x)=-\frac{1}{x^2}$ and not uniformly continuous

### Leipschitz doesn't imply differentiability

Again, we use the example : 
$f(x)=

x & x>0\\
0 & x\leq0
\end{cases}$\
$f$ is K-Liepschitz with $K=1$ but isn't differentiable at 0.

1. Define the function $\Delta(x)$ as
$$  \Delta(x) = \left\{
\begin{array}{ll}
      2x, & x\in [0,0.5) \\      
      2-2x, & x\in [0.5,1) \\
      0, & otherwise \\
\end{array} 
\right.$$
Let $\Delta^2(x)=\Delta(\Delta(x))$, and define $\Delta^3(x)$ etc. similarly. Plot $\Delta, \Delta^2, \Delta^3, \Delta^4$ on $[0,1]$. What is $\Delta^n(x)$.

```python
import numpy as np
import matplotlib.pyplot as plt

    
def Del(x):
    if x>=0 and x<0.5:
        return 2*x
    if x>=0.5 and x<1:
        return 2-2*x
    if x<0 or x>=1:
        return 0
    
def RepeatDel(x,n):
    y=Del(x)
    for i in range(n-1):
        y=Del(y)
    return y

x=np.linspace(0,1,100000)
Delta1=[RepeatDel(z,1) for z in x]
Delta2=[RepeatDel(z,2) for z in x]
Delta3=[RepeatDel(z,3) for z in x]
Delta4=[RepeatDel(z,4) for z in x]
#Delta1000=[RepeatDel(z,60) for z in x]

fig,ax=plt.subplots()
ax.plot(x,Delta1,label='1')
ax.plot(x,Delta2,label='2')
ax.plot(x,Delta3,label='3')
ax.plot(x,Delta4,label='4')
ax.legend()
```

2. With $\epsilon = 0.1$, construct a neural netwrok $f(x)$ with ReLU function to approximate $g(x)=x^2$ in $[-1,1]$. Plot $f(x)$ and $g(x)$ in one figure, and verify that $|f(x)-g(x)|<\epsilon$. 

```python
from math import ceil
import numpy as np
import time
import matplotlib.pyplot as plt
from matplotlib.offsetbox import AnchoredText
start=time.time()

def ReLU(x):
    return max(0,x)


def g(x):
    return x**2


def f(x,alpha,n,a,b):
    s=0
    for i in range(n+1):
        s=s+n*alpha[i]*ReLU((a-x)/(b-a)+(i+1)/n)
    return s
    
a=-1
b=1
e=0.1
n=ceil((b-a)/e)
G = np.linspace(a,b,n+1)
G = np.array([g(x) for x in G]).reshape(n+1,1)




#Construct A
A=np.zeros([n+1,n+1])
for i in range(n-1):
    A[i,i:i+3]=np.array([1,-2,1]).reshape(1,3)    
A[n-1,n-1:n+1]=np.array([1,-2]).reshape(1,2)
A[n,n]=1

#Compute weights
alpha = A@G

#Domain vector
m=50

x = np.linspace(a,b,m)

#Neural Net solutions
y = np.array([f(y,alpha,n,a,b) for y in x]).reshape(m,)

#Function outputs
func = np.array([g(y) for y in x]).reshape(m,)

#Compute error
Error = np.max(np.abs(func-y))
error_text= 'Error is '+str(round(Error,20)) +' < '+'$ \epsilon=$'+str(e)



#Plot results
fig,ax = plt.subplots(figsize=(8,6))

ax.plot(x,func,'rd',label='$g(x)=x^2$')
ax.plot(x,y,'k-',label='Neural net, $f(x)$')
ax.legend(loc=3)
anchored_text = AnchoredText(error_text, loc=9)
ax.add_artist(anchored_text)
ax.set_title('Approximation of $x^2$ with a 1-layer neural network, $\sigma=$ ReLU ')
ax.grid(1)
finish=time.time()
print('Total computation time: ',finish-start)
```

3. Here is the example code for SIR model with actual population $N$.

```python
from scipy.integrate import odeint

# Total population, N. Need to replaced by Arizona population.
N = 1000
# Initial number of infected and recovered individuals, I0 and R0.
I0, R0 = 1, 0
# Everyone else, S0, is susceptible to infection initially.
S0 = N - I0 - R0
# Contact rate, beta, and mean recovery rate, gamma, (in 1/days).
beta, gamma = 0.2, 1./10 
# A grid of time points (in days)
t = np.linspace(0, 160, 160)

# The SIR model differential equations.
def deriv(y, t, N, beta, gamma):
    S, I, R = y
    dSdt = -beta * S * I / N
    dIdt = beta * S * I / N - gamma * I
    dRdt = gamma * I
    return dSdt, dIdt, dRdt

# Initial conditions vector
y0 = S0, I0, R0
# Integrate the SIR equations over the time grid, t.
ret = odeint(deriv, y0, t, args=(N, beta, gamma))
S, I, R = ret.T

plt.plot(t,I)
```
