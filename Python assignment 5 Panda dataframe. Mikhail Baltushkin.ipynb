Write a Python program to draw a scatter plot (similar to the following one) to find sea level rise in 1880-2021. 
Make sure to adjust your figure with the following features,
1. x -axis are labeled by the years;
2. add x and y axies labels;
3. add a title.

Here is the data from [EPA](https://www.epa.gov/climate-indicators/climate-change-indicators-sea-level#:~:text=When%20averaged%20over%20all%20of,as%20the%20long%2Dterm%20trend). Click 'Download Data' under the first figure.

```python
Data = pd.read_csv('sea-level_fig-1.csv',header=None)
```

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(figsize=(8,6))
for i in range(1880,2021):
    x = Data[Data[0]==str(i)]
    ax.plot(float(x[0]),float(x[3]),'.r')
ax.set_title('Rise in Sealevel')
ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (Inches)')
ax.grid(1)
```

## Problem 3

Plot the COVID hospitalization number of Arizona since the fist day in the data spreadsheet.
You may use the data from column 'inpatient_beds_used_covid'.

Download the data from [HealthData.gov](https://healthdata.gov/Hospital/COVID-19-Reported-Patient-Impact-and-Hospital-Capa/g62h-syeh)

In the top right of the website, there is a button to export the csv file.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
Covid=pd.read_csv('COVID-19 data.csv')

a=Covid[Covid['state']=='AZ']

a_sorted = a.sort_values(by ='date')

y=a_sorted['inpatient_beds_used_covid']

time=np.array(a_sorted['date'])

fig,ax = plt.subplots(figsize=(8,6))
ax.plot(time[:180],np.array(y)[:180])
ax.set_xticks(np.concatenate((time[::400],[time[-1]])))
ax.set_title('COVID hospitalizations number of Arizona')
ax.set_xlabel('Date')
ax.set_ylabel('People')
ax.grid(1)
```
